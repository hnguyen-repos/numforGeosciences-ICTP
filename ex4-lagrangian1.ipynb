```python
import numpy as np
import matplotlib.pyplot as plt

# function
def funcAdvec(x):
    y = np.zeros(len(x))
    for i in range(0, len(x)):
        if x[i] < 400 or x[i] > 600:
            y[i] = 0
        elif(x[i] < 500):
            y[i] = 0.1*(x[i] - 400)
        else:
            y[i] = 20 - 0.1*(x[i] - 400)
    return y

def linearInterp(a, y1, y2):
    return (1-a)*y1 + (a)*y2
def cubicInterp(a, x1, x2, x3, x4):
    return -a*(1-a**2)  *x1/6   \
        + a*(1+a)*(2-a) *x2/2   \
        + (1-a**2)*(2-a)*x3/2   \
        - a*(1-a)*(2-a) *x4/6

u = 0.2
dt = 1.0
dx = 5
x0 = np.arange(0, 1000, dx)
y0 = funcAdvec(x0)
xmax = max(x0)
sz = len(x0)
y1 = np.zeros(sz)
tpoint = 1000
t = 0
plt.plot(x0, y0, "+-", label="$\phi$(" + str(int(np.floor(t))) + "s)")
while t < tpoint:
    t = t + dt
    for j in range(sz):
        xjd = x0[j] - u*dt
        if xjd < 0:
            xjd = xjd + xmax
        m = int(np.floor(xjd/dx))
        a = (xjd/dx)-m
        y1[j] = (1-a) * y0[m] + a*y0[m+1]
    y0 = y1
    if t%100 < dt:
        plt.plot(x0, y0,label="$\phi$(" + str(int(np.floor(t))) + "s)")
plt.legend()
plt.grid()
plt.show()
    
```
